```python
bronze_config = [
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_systems/crm/cust_info.csv",
        "table": "crm_cust_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_systems/crm/prd_info.csv",
        "table": "crm_prd_info"
    },
    {
        "source": "crm",
        "path": "/Volumes/workspace/bronze/source_systems/crm/sales_details.csv",
        "table": "crm_sales_details"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_systems/erp/CUST_AZ12.csv",
        "table": "erp_cust_az12"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_systems/erp/LOC_A101.csv",
        "table": "erp_loc_a101"
    },
    {
        "source": "erp",
        "path": "/Volumes/workspace/bronze/source_systems/erp/PX_CAT_G1V2.csv",
        "table": "erp_px_cat_g1v2"
    }
]
```

```python
for item in bronze_config:
    print(f"Copying data from {item['path']} to workspace.bronze.{item['table']}")

    df = (
        spark.read
        .option("header", "true")
        .option("inferSchema", "true")
        .csv(item["path"])
    )
    (
        df.write
        .mode("overwrite")
        .format("delta")
        .saveAsTable(f"workspace.bronze.{item['table']}")
    )
```
