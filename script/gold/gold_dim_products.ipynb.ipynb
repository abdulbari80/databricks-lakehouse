```python
query = """
SELECT
    ROW_NUMBER() OVER(ORDER BY cp.start_date, cp.product_number) AS product_key,
    cp.product_id,
    cp.product_number,
    cp.product_name,
    cp.category_id,
    epc.category,
    epc.subcategory, 
    cp.product_cost,
    epc.maintenance_flag,
    cp.product_line,
    cp.start_date
FROM 
    silver.crm_products AS cp
    LEFT JOIN
        silver.erp_product_category AS epc
    ON cp.category_id = epc.category_id
--WHERE
--    cp.end_date IS NULL
"""
```

```python
df = spark.sql(query)
```

```python
df.limit(5).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.dim_products")
```
