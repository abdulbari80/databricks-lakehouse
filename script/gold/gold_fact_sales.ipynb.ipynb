```python
query = """
SELECT 
    cs.order_number,
    gc.customer_key,
    gp.product_key,
    cs.order_date,
    cs.ship_date,
    cs.due_date,
    cs.price,
    cs.quantity,
    cs.sales_amount
FROM silver.crm_sales as cs
LEFT JOIN gold.dim_customers as gc
    ON cs.customer_id = gc.customer_id 
LEFT JOIN gold.dim_products as gp
    ON cs.product_number = gp.product_number 
"""
```

```python
df = spark.sql(query)
```

```python
df.limit(5).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.gold.fact_sales")
```
