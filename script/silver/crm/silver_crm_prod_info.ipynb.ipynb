```python
import pyspark.sql.functions as F
from pyspark.sql.types import StringType, DateType
from pyspark.sql.functions import col, trim
from pyspark.sql.window import Window
```

```python
df = spark.table("workspace.bronze.crm_prd_info")
```

```python
df.limit(5).display()
```

```python
for field in df.schema.fields:
  if isinstance(field.dataType, StringType):
      df = df.withColumn(field.name, trim(col(field.name)))
```

```python
df = df.withColumn('cat_id', F.regexp_replace(F.substring(col("prd_key"), 1, 5), "-", "_"))
```

```python
df.limit(5).display()
```

```python
df = df.withColumn("prd_key", F.substring(col("prd_key"), 7, F.length(col("prd_key"))))
```

```python
df = df.withColumn("prd_cost", F.coalesce(col("prd_cost"), F.lit(0)))
```

```python
df.select("prd_line").distinct().show()
```

```python
df = (
    df
    .withColumn(
        "prd_line",
        F.when(F.upper(col("prd_line")) == "T", "Touring")
        .when(F.upper(col("prd_line")) == "M", "Mountain")
        .when(F.upper(col("prd_line")) == "S", "Sports")
        .when(F.upper(col("prd_line")) == "R", "Road")
        .otherwise("Unknown")
    )
)
```

```python
df = df.withColumn("prd_start_dt", col("prd_start_dt").cast(DateType()))
```

```python
df.schema["prd_start_dt"].dataType
```

```python
COL_RENAME_MAP = {
    "prd_id": "product_id",
    "cat_id": "category_id",
    "prd_key": "product_number",
    "prd_nm": "product_name",
    "prd_cost": "product_cost",
    "prd_line": "product_line",
    "prd_start_dt": "start_date",
    "prd_end_dt": "end_date"
}
for old_name, new_name in COL_RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```

```python
df.limit(5).display()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.crm_products")
```
